```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,5,9],
                   'Col_1':[1,2,3,4,5],
                   'Col_2':[6,7,8,9,10],
                   'Col_3':[11,12,13,14,15],
                   'Col_4':['apple','orange','banana','strawberry','raspberry']})
df2 = pd.DataFrame({'ID':[1,1,3,5],
                    'Col_A':[8,9,10,11],
                    'Col_B':[12,13,15,17],
                    'Col_4':['apple','orange','banana','kiwi']})
```

```python
df1
```

```python
df2
```

```python

pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on='ID')
```

```python
pd.merge(df1,df2, on=['ID','Col_4'])
#this is the same as the first merge beacuse we are specifying alll colums that share a name between the two DataFrames
```

```python
pd.merge(df1,df2, suffixes = ['_h','_b'],left_on = 'Col_2', right_on = 'Col_A' )
```

```python
pd.merge(df1,df2, suffixes = ['_h','_i'],left_index = True, right_index = True )
```

```python
pd.merge(df1,df2, suffixes = ['_h','_i'], on = 'ID' )
```

```python
#Outer join
pd.merge(df1,df2, on='Col_4', how='outer')
```

```python
#Left join
pd.merge(df1,df2, on='Col_4', how='left')
```

```python
#Right join
pd.merge(df1,df2, on='Col_4', how='right')
```

```python
#merge is better than join --- Side by side 
#Concat is better than append --- Stack on top/below of each other
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis = 1)
#become like Merge
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
pd.concat([df1, df2], axis = 0, join = 'inner')
```

```python
df1.append(df2, sort=True)
```
